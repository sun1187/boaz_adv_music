# 2) **1방법 + 가중치**

- 노래마다 tag들의 개수를 구한다. ex - 노래 {tag1 : 3, tag2 : 3, tag3 :4}
- tag마다 가중치를 구함 - ex 해당 tag개수 / 총 개수 (softmax 느낌)
- 나온 가중치를 통해 tag들이 속한 감정을 label로 구함(ex - 우울한 0.2, 신나는 0.4 ...)
- nlp에서 output을 낼 때 가중치로 여러개 냄(softmax) → 몇 개로??
- nlp의 가중치 output과 tag들의 가중치 output에 가장 가까운 노래를 추천(유클리디안 거리, 코사인 등)

### 태그별 가중치
__태그별 나오는 횟수를 기준으로 전체에서 차지하는 비율 계산__

<br>

- 원본: ['비오는날', '드라이브', '비오는날', '추억', '회상']
- 중복 없는 리스트 ['비오는날', '추억', '드라이브', '회상']
- 각 태그가 나오는 횟수 [2, 1, 1, 1]
- 각 태그가 차지하는 비율 [0.4, 0.2, 0.2, 0.2] 

```python
import numpy as np
from collections import Counter

def softmax(t):
    s = set(t)
    c = [dict(Counter(t))[name] for name in s]
    return dict(zip(s, np.array(c) / sum(c)))

softmax(df['tag'][0].split(', '))
```

```python
print(df['tag'][1].split(', '))
softmax(df['tag'][1].split(', '))
```

```python
print(df['emotion'][1].split(', '))
softmax(df['emotion'][1].split(', '))
```

```python
df['weight'] = pd.Series([softmax(df['emotion'][i].split(', ')) for i in range(len(df))]) 
df
```
