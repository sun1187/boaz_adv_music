# 1. 가사 기반 TFIDF_Cosine

```python
from sklearn.metrics.pairwise import cosine_similarity

text = "어디든 괜찮을 것 같은데? 하아 힘든 세상.어디 하나 기댈 데도 없는.이 세상.너뿐이다. 아름다워. 삶은 아름답지만 가끔 도망치는 것도 괜찮을 것 같은데? 두근거렸지."
text_matrix = tfidfv.transform([text])
```

```python
cos_list = []

for i in range(len(lyrics_matrix)):
    cos = cosine_similarity(text_matrix, lyrics_matrix[i:i+1])
    cos_list.append(cos[0][0])
```

```python
n = 5
aa = np.array(cos_list).argsort()[-n:][::-1]

pd.DataFrame(df.iloc[aa])[['가수', '노래', '원가사']]
```

# 2. 가사+태그 기반_TFIDF_Cosin

```python
text = "어디든 괜찮을 것 같은데? 하아 힘든 세상.어디 하나 기댈 데도 없는.이 세상.너뿐이다. 아름다워. 삶은 아름답지만 가끔 도망치는 것도 괜찮을 것 같은데? 두근거렸지."

check_tag = '위로'
# 해당 태그가 있는 노래들로 1차 필터링
check_tag_df = df[df.tag.apply(lambda x:check_tag in x)]
```

```python
# 1차 필터링 중 글과 유사한 가사 순으로 보여주기
from sklearn.feature_extraction.text import TfidfVectorizer

tfidfv = TfidfVectorizer()
tfidf_matrix = tfidfv.fit_transform(check_tag_df['1차 전처리 후'].values)
text_matrix = tfidfv.transform([text])

lyrics_matrix = pd.DataFrame(tfidf_matrix.toarray(), columns = tfidfv.get_feature_names())

cos_list = []

for i in range(len(lyrics_matrix)):
    cos = cosine_similarity(text_matrix, lyrics_matrix[i:i+1])
    cos_list.append(cos[0][0])
    
n = 5
aa = np.array(cos_list).argsort()[-n:][::-1]
check_tag_df['cos'] = cos_list
pd.DataFrame(check_tag_df.iloc[aa])[['가수', '노래', '원가사', 'cos']]
```
