```python
## 추천 구하기
import warnings
warnings.filterwarnings('ignore')
from sklearn.metrics.pairwise import cosine_similarity

rec_result = []

for j in tqdm(range(len(new_test_df))):
    target = new_test_df['tag_vec_ft'].values[j][0].reshape(1, -1)
    
    train_df['cos_sim'] = train_df['tag_vec_ft'].apply(lambda _df: cosine_similarity(_df, target))

    rec_result.append((train_df.iloc[np.argmax(train_df['cos_sim'])]['곡 리스트'],
                      np.argmax(train_df['cos_sim']), max(train_df['cos_sim'])))
    
```

```python
# 평가하기
class Evaluator:
    def _idcg(self, l):
        return sum((1.0 / np.log(i + 2) for i in range(l)))

    def __init__(self):
        self._idcgs = [self._idcg(i) for i in range(101)]

    def _ndcg(self, gt, rec):
        dcg = 0.0
        for i, r in enumerate(rec):
            if r in gt:
                dcg += 1.0 / np.log(i + 2)
        #print(len(gt))
        #print(len(rec))
        #print()
        return dcg / self._idcgs[len(gt)]
    
evaluator = Evaluator()
```

```python
ndcg_total_score = []

for i in tqdm(range(len(new_test_df))):
    num = i

    true_relevance = new_test_df['곡 리스트'][num][:100]
    scores = rec_result[num][0][:100]
            
    ndcg_total_score.append(evaluator._ndcg(true_relevance, scores))    
print(np.mean(ndcg_total_score))
```

```python
# 유저 관점:

user_tag = '힐링, 쉼, 휴식, 고요'
emotion_result = ('잔잔한', 0.7), ('편안한', 0.3)
top_n = 5
max_song_list = 100
################

emotion_n = len(emotion_result)

sa_active = 0
h_active = 0
p_active = 0
l_active = 0
sc_active = 0
co_active = 0
cr_active = 0
n_active = 0

for emo_idx in range(len(emotion_result)):
    emo = emotion_result[emo_idx][0]
    if emo == '잔잔한':
        p_active = emotion_result[emo_idx][1]
    if emo == '편안한':
        co_active = emotion_result[emo_idx][1]
    if emo == '우울한':
        sa_active = emotion_result[emo_idx][1]
    if emo == '행복한':
        h_active = emotion_result[emo_idx][1]
    if emo == '경쾌한':
        l_active = emotion_result[emo_idx][1]
    if emo == '무서운':
        sc_active = emotion_result[emo_idx][1]
    if emo == '울고싶은':
        cr_active = emotion_result[emo_idx][1]
    if emo == '긴장되는':
        n_active = emotion_result[emo_idx][1]
```

```python
import heapq
fr = train_df['cos_sim']
n_result = heapq.nlargest(max_song_list, range(len(fr)), key=fr.__getitem__)

print('active 값:', sa_active, h_active, p_active, l_active, sc_active, co_active, cr_active, n_active)

found_list =[]
found = 0
flag = 1

for i in range(max_song_list):
    sc_list = []
    rec_result = []
    rec_result.append((train_df.iloc[n_result[i]]['곡 리스트'],
                   train_df.iloc[n_result[i]]['cos_sim']))

    user_rec_songs = rec_result[0][0].replace('[', '').replace(']', '').replace("'", '').replace(" ", '').split(',')#[:100]
    user_rec_songs = list(map(int, user_rec_songs))
    print('similarity', rec_result[0][1])

    for song in user_rec_songs:
        sc = 0
        if sa_active:
            if song in sad:
                sc += sa_active
        if h_active:
            if song in happy:
                sc += h_active
        if p_active:
            if song in peace:
                sc += p_active
        if l_active:
            if song in light:
                sc += l_active
        if sc_active:
            if song in scary:
                sc += sc_active
        if co_active:
            if song in comfort:
                sc += co_active
        if cr_active:
            if song in cry:
                sc += cr_active
        if n_active:
            if song in nerbous:
                sc += n_active
        
        if sc>0:
            print('song:', song, 'sc:', sc)
            sc_list.append(sc)

            found += 1
            found_list.append(song)
        
    if found >= top_n:
        break
```
