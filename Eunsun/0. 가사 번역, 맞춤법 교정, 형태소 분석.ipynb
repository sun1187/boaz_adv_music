```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('genie 크롤링 후_681.csv')
data.drop('Unnamed: 0', axis=1, inplace=True)
data
```

```python
# 영어 스펠링 체크 > 한국어로 번역 > 한국어 맞춤법 교정
from tqdm import tqdm
from hanspell import spell_checker
from gingerit.gingerit import GingerIt
import math

import googletrans
translator = googletrans.Translator()
finish = []
sub_data = data.copy()
sub_data = sub_data[155:]

for i in tqdm(range(155, len(data))):
    result = []
    ori_target = sub_data['원가사'][i]
    lyric_length = len(ori_target)
    need_list = math.ceil(lyric_length / 399) # 1.4개의 리스트가 필요하면 2개가 필요하다고 본다.
    #print(lyric_length)
    #print(need_list)
    for i in range(need_list):
        # 영어 스펠링 체크
        target = ori_target[i*399:(i+1)*399]
        target = target.replace("\'", "")
        target = target.replace("&", " ")
      #  print(target)
      #  print()
        try:
            corrected_text1 = GingerIt().parse(target)
            to_kor = corrected_text1['result']
     #       print(11)
        except:
            to_kor = target
      #      print(22)
        
       # print('to_kor', to_kor)
        # 한국어로 번역하기
        translated = translator.translate(to_kor, dest='ko').text
        #print(translated)
        #print()
        #print('!번역!', translated)
        
        # 한국어 맞춤법 교정
        result.append(spell_checker.check(translated).checked)
        #print()
        #print('교정!', result)
        #print()
    finish.append(",".join(result))
    
```

```python
sub3 = data[155:155+len(finish)]
sub3['1차 가공'] = finish
sub3.to_csv('가공_3.csv', encoding='utf-8-sig')
```

```python
df1 = pd.read_csv('가공_1.csv')
df1.columns = ['index_n', '노래', '가수', '원가사', '좋아요수', '어제순위', '댓글1', '댓글2', '댓글3',
       '댓글4', '댓글5', '1차 가공']
df1.set_index('index_n', inplace=True, drop=True)

df2 = pd.read_csv('가공_2.csv')
df2.columns = ['index_n', '노래', '가수', '원가사', '좋아요수', '어제순위', '댓글1', '댓글2', '댓글3',
       '댓글4', '댓글5', '1차 가공']
df2.set_index('index_n', inplace=True, drop=True)

df3 = pd.read_csv('가공_3.csv')
df3.columns = ['index_n', '노래', '가수', '원가사', '좋아요수', '어제순위', '댓글1', '댓글2', '댓글3',
       '댓글4', '댓글5', '1차 가공']
df3.set_index('index_n', inplace=True, drop=True)
```

```python
df_all = pd.concat([df1, df2, df3], axis=0)
df_all.head()
```

```python
# 예시
from konlpy.tag import Okt

voc = '형태소 란 문장을 구성하는 너무 매우 의미 요소 중 가장 작은 단위를 말한다.'
okt_pos = Okt().pos(voc, norm=True, stem=True)   # 형태소 분석 ( norm : 정규화 )


okt_filtering = [x for x, y in okt_pos if y in ['Noun', 'Adjective', 'Verb', 'Adverb']]
print(okt_filtering)   
```

```python
from konlpy.tag import Okt

def okt_organize(text):
    okt_pos = Okt().pos(text, norm=True, stem=True)
    okt_filtering = [x for x, y in okt_pos if y in ['Noun', 'Adjective', 'Verb', 'Adverb']]
    return okt_filtering
```

```python
import re

def remove_dirty(text):
    result = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]', '', text)

    return result
```

```python
df_all['2차 가공'] = df_all['1차 가공'].apply(lambda x: remove_dirty(x))
```

```python
df_all['3차 가공'] = df_all['2차 가공'].apply(lambda x: okt_organize(x))
```
