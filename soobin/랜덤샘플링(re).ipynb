```python
import re
 
def cleanText(readData):
 
    #텍스트에 포함되어 있는 특수 문자 제거
 
    text = re.sub('[-=+#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]', '', readData)
 
    return text
```

```python
gr = song['곡 장르 리스트']
sub = []
for n,i in enumerate(gr):
  a = cleanText(str(i))
  sub.append(a)

sub
```

```python
gr = song['곡 세부 장르 리스트']
sub2 = []
for n,i in enumerate(gr):
  a = cleanText(str(i))
  sub2.append(a)

sub2
```

```python
import pandas as pd
genre_new = pd.read_excel('/content/drive/MyDrive/adv/장르_전처리_최리.xlsx',sheet_name='장르 전처리')
genre_new
```

```python
genre_new['대분류'].unique()
```

```python
big_cat_gen2 = ['GN0100', 'GN0200', 'GN0300', 'GN1200', 'GN0400', 'GN1300',
       'GN0500', 'GN0600', 'GN1000', 'GN0900', 'GN1100', 'GN2600',
       'GN1400', 'GN0800', 'GN1600', 'GN1700', 'GN1800', 'GN2500',
       'GN2700', 'GN2800', 'GN2900', 'GN3000']
```

```python
new_so = list(genre_new['소분류'])
new_so  
```

```python
index = []
for i in range(len(song)):
  a = song['소분류str'][i].split(',')
  for j in a:
    if j in new_so:
      index.append(i)
      break
index
```

```python
len(index)
```

```python
new_song = song.iloc[index]  #세부장르 거른거  (우리가 선택한 소분류에 속하는 데이터만 뽑아서 new_song에 저장함)
new_song
```

```python
big_cat_gen2
```

```python
ans = []
for i in big_cat_gen2:
  if i == 'GN3000':
    break
  print(i)
  if i in ['GN0300','GN1200','GN0400','GN1300','GN0600','GN1000','GN1100','GN2600','GN1400','GN0800']:
    sub = new_song[new_song['대분류str'].str.contains(i)]
    sub = list(sub.sample(50,random_state=924).index)
    for i in sub:
      ans.append(i)
  else:
    sub = new_song[new_song['대분류str'].str.contains(i)]
    sub = list(sub.sample(100,random_state=924).index)
    for i in sub:
      ans.append(i)
  
```

```python
print(len(new_song))
print(ans[:10])
```

```python
new_song = new_song.set_index("index")
new_song
```

```python
sample_song = new_song.loc[ans]
sample_song
```

```python
from tqdm import tqdm  # 대분류 합치기로 한거 합치기 
import time

for i in tqdm(range(len(sample_song)),desc="tqdm example"):
  a = sample_song['대분류str'].iloc[i].split(',')
  sub = []
  for j in a:
    j=j.strip()
    if j=='GN1200':
      sub.append('GN0300')
    elif j=='GN1300':
      sub.append('GN0400')
    elif j=='GN1000':
      sub.append('GN0600')
    elif j=='GN2600':
      sub.append('GN1100')
    elif j=='GN0800':
      sub.append('GN1400')
    else:
      sub.append(j)

  fin = ",".join(sub)
  sample_song['대분류str'].iloc[i] = fin
  
sample_song
```

```python
def get_tags(tag_list, threshold=0):
  if len(tag_list) > 0:
    return ' '.join([x for x in tag_list])
  else:
    return ''


train['tags_preprocessed'] = train['tags'].apply(lambda x: get_tags(x))

train
```

```python
from tqdm import tqdm
import time

result = ['']*len(song)
for i in tqdm(range(len(train)),desc="tqdm example"):
  for j in train['songs'][i]:
    if j in ans:
      result[j]+=train['tags_preprocessed'][i]
      # print(result[j])
```

```python
result_pd = pd.DataFrame(result,columns=['tags'])
```

```python
len(result_pd)
```

```python
result_pd = result_pd[result_pd['tags']!='']   #tag있는 경우만 뽑아서 저장
result_pd.reset_index(inplace=True)
result_pd
```
