```python
import pandas as pd
sample_song_tag = pd.read_csv('sample_song_tag(re).csv')
sample_song_tag
```

```python
df = sample_song_tag[['곡 제목', '곡 ID', '대분류str', '소분류str', 'tags']]
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vector = CountVectorizer()
```

```python
df.dropna(axis=0, inplace=True)
```

```python
df.isnull().sum()
```

```python
count_vec = vector.fit_transform(df['tags']).toarray()
print(vector.vocabulary_)
```

```python
vec = pd.DataFrame(data = count_vec)
```

```python
vec.head()
```

```python
# vec['vec_sum'] = vec.sum()
# vec.head()
```

```python
vec_t = vec.T
vec_t
```

```python
vec_max = vec_t.idxmax()
```

```python
vec['max_idx'] = vec_max
```

```python
vec.head()
```

```python
vector_dict = vector.vocabulary_
```

```python
dict_key = list(vector_dict.keys())
dict_value = list(vector_dict.values())
```

```python
vec['max_idx'][0]
```

```python
key_list = []

for i in range(len(vec)):
    idx_value = vec['max_idx'][i]
    key = dict_key[dict_value.index(idx_value)]
    key_list.append(key)
```

```python
vec['max_key'] = key_list
```

```python
max_key = vec['max_key']
```

```python
max_key
```

```python
df_max = pd.concat([df, max_key], axis=1)
```

```python
max_key
```

```python
df_max
# 개수 왜 안 맞지?
```

- 위처럼 개수가 큰 애들순으로 구하기
- 가중치 구하기
- 가중치를 바탕으로 tag -> label 매칭

# 가중치 추가

```python
vec_drop = vec.drop(['max_idx', 'max_key'], axis=1)
```

```python
vec_sum = vec_drop.sum(axis=1)
vec_max = vec_drop.max(axis=1)
```

```python
vec_max_prob = vec_max/vec_sum
```

```python
df_max['max_prob'] = vec_max_prob
```
